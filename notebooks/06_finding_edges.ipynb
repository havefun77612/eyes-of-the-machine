# Chapter 6: Finding the Edges
## Companion Notebook — "Eyes of the Machine"

Learn to detect edges — the first step the machine takes toward actually seeing shapes.

---
### 6A: Setup — Imports and Image Load

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

img_bgr = cv.imread(filename)
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img_bgr, cv.COLOR_BGR2GRAY)

plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')
plt.show()
```

---
### 6B: Canny Edge Detection
The gold standard. Finds sharp brightness changes.

```python
edges = cv.Canny(img_gray, 50, 150)

print("Edge image shape:", edges.shape)
print("White pixels = edges, Black = no edge")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edges (50, 150)")
plt.axis('off')
plt.show()
```

---
### 6C: Experimenting with Thresholds
Low thresholds = more edges. High thresholds = only strongest edges.

```python
edges_loose = cv.Canny(img_gray, 10, 50)
edges_tight = cv.Canny(img_gray, 100, 200)
edges_balanced = cv.Canny(img_gray, 50, 150)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(edges_loose, cmap='gray')
plt.title("Loose (10, 50)")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edges_balanced, cmap='gray')
plt.title("Balanced (50, 150)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges_tight, cmap='gray')
plt.title("Tight (100, 200)")
plt.axis('off')
plt.show()
```

---
### 6D: Sobel — Directional Edge Detection
Detect edges in a specific direction (horizontal or vertical).

```python
sobel_x = cv.Sobel(img_gray, cv.CV_64F, 1, 0, ksize=5)
sobel_y = cv.Sobel(img_gray, cv.CV_64F, 0, 1, ksize=5)

sobel_x_display = np.uint8(np.abs(sobel_x))
sobel_y_display = np.uint8(np.abs(sobel_y))
sobel_combined = cv.bitwise_or(np.abs(sobel_x), np.abs(sobel_y))

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(sobel_x_display, cmap='gray')
plt.title("Sobel X (Horizontal edges)")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(sobel_y_display, cmap='gray')
plt.title("Sobel Y (Vertical edges)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobel_combined, cmap='gray')
plt.title("Sobel Combined")
plt.axis('off')
plt.show()
```

---
### 6E: Blur Before Edge Detection
Blur removes noise so only real object boundaries survive.

```python
blurred = cv.GaussianBlur(img_gray, (5, 5), 0)
edges_clean = cv.Canny(blurred, 50, 150)
edges_no_blur = cv.Canny(img_gray, 50, 150)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(edges_no_blur, cmap='gray')
plt.title("Edges without blur — noisy")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges_clean, cmap='gray')
plt.title("Edges with blur — cleaner")
plt.axis('off')
plt.show()
```

---
### 6F: Edge Overlay — Green Edges on Original
See where the edges fall on the original image.

```python
overlay = img_rgb.copy()
overlay[edges_clean > 0] = [0, 255, 0]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(overlay)
plt.title("Edges Overlayed (Green)")
plt.axis('off')
plt.show()
```

---
### 6G: Pencil Sketch Effect
Invert the edge image for a drawing-like result.

```python
sketch = cv.bitwise_not(edges_clean)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Photo")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sketch, cmap='gray')
plt.title("Pencil Sketch Effect")
plt.axis('off')
plt.show()
```

## 🧠 Challenge: The "Roommate Detection" Messiness Counter

Take a photo of a messy space. Write code that:
1. Counts edges with Canny.
2. Prints the edge count as a "messiness score."
3. Blurs heavily and counts edges again ("after cleaning").
4. Displays all three side by side.

```python
# Count edges.
edges_messy = cv.Canny(img_gray, 50, 150)
edge_count = np.count_nonzero(edges_messy)
print(f"Your space has {edge_count} edges of messiness.")

# "Clean" by blurring heavily.
blurred_heavy = cv.GaussianBlur(img_gray, (51, 51), 0)
edges_clean_room = cv.Canny(blurred_heavy, 50, 150)
clean_count = np.count_nonzero(edges_clean_room)
print(f"After cleaning: only {clean_count} edges remain.")

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(edges_messy, cmap='gray')
plt.title(f"Messy: {edge_count} edges")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(edges_clean_room, cmap='gray')
plt.title(f"Clean: {clean_count} edges")
plt.axis('off')
plt.show()
```
