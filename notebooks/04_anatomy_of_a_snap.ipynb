### 4A: Creating an Image from Scratch (Numbers → Picture)
An image is just a numpy array of numbers. Let's prove it.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 50x50 red image.
red_image = np.zeros((50, 50, 3), dtype=np.uint8)
red_image[:, :] = [255, 0, 0]

plt.imshow(red_image)
plt.title("Made of Numbers: Red (255, 0, 0)")
plt.axis('off')
plt.show()
```

---
### 4B: Making a Gradient (Seeing the Numbers Change)
A smooth transition from black to blue shows pixels changing value gradually.

```python
gradient = np.zeros((100, 256, 3), dtype=np.uint8)

for i in range(256):
    gradient[:, i, 2] = i

plt.imshow(gradient)
plt.title("Gradient: Black to Blue (Blue channel 0 to 255)")
plt.axis('off')
plt.show()
```

### 4D: Inspecting the Image's Structure
Let's look at the shape, data type, and visual.

```python
image = plt.imread(filename)

print("Image shape:", image.shape)
print("Data type:", image.dtype)
print("Height (rows):", image.shape[0])
print("Width (cols):", image.shape[1])
print("Channels:", image.shape[2] if len(image.shape) == 3 else 1)

plt.imshow(image)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

---
### 4E: Peeking at the Numbers
An image is JUST a grid of numbers. Let's see them.

```python
print("First row, first 10 pixels of RED channel:")
print(image[0, :10, 0])

print("First row, first 10 pixels of GREEN channel:")
print(image[0, :10, 1])

print("First row, first 10 pixels of BLUE channel:")
print(image[0, :10, 2])

print("\nPixel at center (row 100, col 100):", image[100, 100])
```

---
### 4F: Manipulating Pixels Directly
Modify the image by changing the numbers in the array.

```python
modified = image.copy()
modified[20:60, 30:70] = [0, 0, 1.0]

plt.imshow(modified)
plt.title("Blue Square Painted on Image")
plt.axis('off')
plt.show()
```

---
### 4G: Converting to Grayscale
Reduce 3 channels to 1 by averaging brightness.

```python
grayscale = np.mean(image, axis=2)

print("Original shape:", image.shape)
print("Grayscale shape:", grayscale.shape)

plt.imshow(grayscale, cmap='gray')
plt.title("Grayscale: Every Pixel = Brightness Only")
plt.axis('off')
plt.show()
```
