### 3A: Variables — Your First Sticky Notes
A variable stores a value in the computer's memory, like a labeled sticky note on a wall.

```python
# Store values in variables (labeled sticky notes).
student_name = "Aisha"
student_age = 19
student_height = 1.65
is_beginner = True

# Print displays the value of each variable.
print("Name:", student_name)
print("Age:", student_age)
print("Height (m):", student_height)
print("Beginner?:", is_beginner)
```

---
### 3B: Changing Variables (Overwriting)
Variables can be reassigned. The old value is replaced.

```python
# Assign and then overwrite.
mood = "tired"
print("Morning mood:", mood)

mood = "awake"
print("After coffee mood:", mood)

# Update using the old value.
coins_in_pocket = 5
print("Before lunch:", coins_in_pocket)

coins_in_pocket = coins_in_pocket + 3
print("After finding coins:", coins_in_pocket)

# Shortcut: += adds to the current value.
coins_in_pocket += 2
print("After another miracle:", coins_in_pocket)
```

### 3C: Lists — The Shopping Bag
A list holds multiple items in order. Access items by index (starting at 0).

```python
# Create a list with square brackets.
subjects = ["Math", "Physics", "Computer Science", "Psychology"]
print("My subjects:", subjects)

# Access by index (starts at 0!).
first = subjects[0]
print("First (index 0):", first)

second = subjects[1]
print("Second (index 1):", second)

# Index -1 = last item.
last = subjects[-1]
print("Last (index -1):", last)

# Count items with len().
print("Count:", len(subjects))

# Add an item with append().
subjects.append("Art")
print("After adding Art:", subjects)
```

### 3D: If/Else — The Decision Maker
Make decisions with yes/no questions.

```python
# Compare values with == (two equals signs).
temperature = 30

if temperature > 25:
    print("It's hot! Wear a t-shirt.")
else:
    print("It's cool. Bring a jacket.")

# Multiple conditions with elif.
grade = 85

if grade >= 90:
    print("You got an A!")
elif grade >= 80:
    print("You got a B.")
elif grade >= 70:
    print("You got a C.")
else:
    print("Keep trying.")
```

### 3E: For Loops — The Repeater
Repeat code for each item in a list.

```python
# Loop through a list.
subjects = ["Math", "Physics", "CS", "Psychology"]

for subject in subjects:
    print("I have class in:", subject)

print("---")

# Loop with range().
for number in range(5):
    print("Iteration:", number)

print("---")

# Loop with index using enumerate.
for index, subject in enumerate(subjects):
    print(index, "→", subject)
```

### 3F: Functions — The Recipe
A function is a reusable block of code. Define with 'def', use with a call.

```python
# Define a function.
def greet(name):
    message = "Hello, " + name + "!"
    return message

# Call the function.
result = greet("Hossam")
print(result)

# Function with two inputs.
def make_juice(fruit, count):
    glasses = fruit * count
    return str(glasses) + " glasses of juice"

print(make_juice("orange", 3))
```

### 3G: Putting It All Together — The "Smart Morning" Script
A complete script using variables, conditions, lists, and loops.

```python
# Rate your energy from 1 to 5.
energy_level = 3
tasks = ["Shower", "Eat", "Study", "Exercise"]

if energy_level <= 2:
    print("You need coffee. And maybe a nap.")
elif energy_level <= 4:
    print("Manageable. Let's go.")
else:
    print("You're a superhero. Teach a class.")

for task in tasks:
    print("✓ Doing:", task)

print("Morning complete. You survived.")
```
